```python
import pandas as pd
```

```python
sr = pd.read_csv("size_results.csv")
cifar = sr[sr["Dataset"]=="CIFAR-10"]
fem = sr[sr["Dataset"]=="FEMNIST"]
```

```python
top_fem = fem.sort_values("Data Uploaded").groupby("Goal Accuracy").head(3).sort_values("Goal Accuracy").reset_index(drop=True)
old_idx = top_fem.index.to_frame()
old_idx.insert(0, "Goal Accuracy", top_fem["Goal Accuracy"])
top_fem.index = pd.MultiIndex.from_frame(old_idx)
top_fem.drop("Goal Accuracy", axis=1)
```

```python
top_cifar = cifar.sort_values("Data Uploaded").groupby("Goal Accuracy").head(3).sort_values("Goal Accuracy").reset_index(drop=True)
old_idx = top_cifar.index.to_frame()
old_idx.insert(0, "Goal Accuracy", top_cifar["Goal Accuracy"])
top_cifar.index = pd.MultiIndex.from_frame(old_idx)
top_cifar.drop("Goal Accuracy", axis=1)
```
